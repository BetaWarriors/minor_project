```python
X = nasdaq_wkly_df.drop(["bins","target", "Sa", "Su", "weeks"], axis  = 1)
Y = nasdaq_wkly_df.bins
X, Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 60)
```

```python
X_train, X_test, Y_train, Y_test
```

```python
%%time
x = float('inf')
import numpy as np
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, Y_train, sample_weight=[2 for i in range(len(X_train))])
```

```python
model.score(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred.shape, Y_test.shape
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, y_pred), model.score(X_test,Y_test)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=2,figsize=(15,20))
ax[0].plot(list(Y_test), label="Actual Stocks")
ax[1].plot(y_pred, label = "predicted Stocks")
ax[0].legend()
ax[1].legend()
```
